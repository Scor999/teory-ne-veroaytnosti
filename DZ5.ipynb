**1. Когда используется критерий Стьюдента, а когда Z –критерий?**

- Если условия применимости параметрических тестов соблюдены, то в случае известной сигмы генеральной совокупности более, предпочтителен Z- критерий.
- Особенности использования t-критерия Стьюдента. Наиболее часто t -критерий используется в двух случаях. 
  - В первом случае его применяют для проверки гипотезы о равенстве генеральных средних двух независимых, несвязанных выборок (так называемый двухвыборочный t-критерий).

  - Во втором же случае используется так называемый парный t-критерий, когда одна и та же группа объектов порождает числовой материал для проверки гипотез о средних. Поэтому эти выборки называют зависимыми, связанными.

**2. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с $\alpha=0.05$ проверить эту гипотезу, если в выборке из $n=100$ шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.**

Формулируем гипотезу $H0$:  
$H0: \mu=\mu_0$                                                  
Формулируем гипотезу $H1$:                                      
$H1: \mu>\mu_0$  
Поскольку нам известны и среднее и дисперсия генеральной совокупности, то для проверки гипотезы будем использовать $z$-критерий.  
$z_н=\frac{17,5-17}{2/\sqrt 100}=2.5$               
Табличное значение $z$-критерия, для $\alpha=0.05$ равно $z_t$= 1,645                   
Так как $z_н$>$z_t$, то верна гипотеза $H1$.



**3.Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?**

Формулируем гипотезу $H0$:  
$H0: \mu=\mu_0$                                                 
Формулируем гипотезу $H1$:                                      
$H1: \mu\not=\mu_0$             
Поскольку мы проверяем гипотезу относительно математического ожидания нормально распределённой случайной величины с неизвестной дисперсией, будем использовать статистику:               
$|t_н| = \dfrac{\overline{X} - a}{\sigma_X / \sqrt{n}}= \dfrac{\overline{X} - 200}{\sigma_X / \sqrt{10}} = \dfrac{\sqrt{10} \cdot (198.5 - 200)}{4.45}= -1.0659$                 
Найдем табличное значение $t$-критерия $t_t=3.2498$         
Так как $|t_н|<t_t$, то **принимаем** нулевую гипотезу на уровне значимости $\alpha=0.01$.

**4.Есть ли статистически значимые различия в росте дочерей?**                
**Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165**            
**Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160, 163**

```python
import numpy
from scipy import stats
import matplotlib.pyplot as plt


mothers = numpy.array([172, 177, 158, 170, 178, 175, 164, 160, 169, 165])

daughters = numpy.array([173, 175, 162, 174, 175, 168, 155, 170, 160, 163])

# plt.bar(numpy.arange(10) - 0.2, mothers, width=0.3, label='Рост матерей')
# plt.bar(numpy.arange(10) + 0.2, daughters, width=0.3, label='Рост дочерей')
# plt.xticks(numpy.arange(10))
# plt.legend()
# plt.title('Сравнительный график роста матерей и дочерей')


stats.ttest_rel(mothers, daughters)
```

Так как $p-value(0,5228) > \alpha (0,05)$ , то статистически значимых различий в росте дочерей нет.
